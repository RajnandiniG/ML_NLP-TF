```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Read the image
img = cv2.imread('/content/house.jpg')

# Convert BGR (OpenCV default) to RGB (matplotlib default)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/content/newtype.jpg')

# Convert the image from BGR to RGB, for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Read class labels
all_rows = open('/content/synset_words.txt').read().strip().split("\n")
classes = [r[r.find(' ') + 1:] for r in all_rows]

# Print the first 5 classes
for i, c in enumerate(classes):
    if i == 5:  # Changed to print the first 5 classes
        break
    print(i, c)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/content/typewriter.jpg')

# Convert the image from BGR to RGB, for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

net = cv2.dnn.readNetFromCaffe('/content/bvlc_googlenet.prototxt','/content/bvlc_googlenet.caffemodel')

blob = cv2.dnn.blobFromImage(img, 1, (224,224))

net.setInput(blob)

outp = net.forward()

idx = np.argsort(outp[0])[::-1][:5]

for (i,id) in enumerate(idx):
    print('{}. {} ({}): Probability {:.3}%'.format(i+1, classes[id], id, outp[0][id]*100))

plt.imshow(img_rgb)
plt.axis('off')  # Hide the axis
plt.show()
```
